```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Specify the ticker symbol of the stock you want to fetch data for
ticker_symbol = 'AAPL'

# Connect to the Yahoo Finance API and fetch historical stock data
stock_data = yf.Ticker(ticker_symbol)

# Get historical stock data for the specified ticker symbol
historical_data = stock_data.history(period='1y')

# Plot the closing price of the stock over one year
plt.figure(figsize=(12, 6))
plt.plot(historical_data.index, historical_data['Close'], label='Closing Price', color='blue')

# Add a trendline (linear regression)
x = np.arange(len(historical_data))
coefficients = np.polyfit(x, historical_data['Close'], 1)
trendline = np.polyval(coefficients, x)
plt.plot(historical_data.index, trendline, label='Trendline', color='red')

plt.title(f'{ticker_symbol} Stock Price Over 1 Year with Trendline')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Specify the ticker symbols of the stocks you want to fetch data for
ticker_symbols = ['AAPL', 'GLE.PA']
colors = ['blue', 'orange']

# Create an empty DataFrame to store the historical stock data
historical_data = pd.DataFrame()

# Fetch historical stock data for each ticker symbol and store it in the DataFrame
for i, ticker_symbol in enumerate(ticker_symbols):
    stock_data = yf.Ticker(ticker_symbol)
    stock_prices = stock_data.history(period='max')['Close']
    historical_data[ticker_symbol] = stock_prices

# Plot the maximum historical stock prices of the stocks
plt.figure(figsize=(12, 6))
for i, ticker_symbol in enumerate(ticker_symbols):
    plt.plot(historical_data.index, historical_data[ticker_symbol], label=ticker_symbol + ' Max Price', color=colors[i])

plt.title('Maximum Historical Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
